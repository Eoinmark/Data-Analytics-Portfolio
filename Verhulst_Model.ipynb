```python
import matplotlib.pyplot as plt #will be used for plotting
import numpy as np

# Verhulst's logistic growth model parameters
r = 0.2 # growth rate
K = 1000 # carrying capacity
N = 50 # initial population size

# create empty arrays to store time and population size values
t = np.zeros(101)
N_vals = np.zeros(101)

# set initial values
t[0] = 0
N_vals[0] = N

# simulate population growth using logistic growth equation and Euler method
for i in range(1, 101):
    t[i] = i
    N_vals[i] = N_vals[i-1] + r*N_vals[i-1]*(1 - N_vals[i-1]/K)
    
# plot population growth over time
plt.plot(t, N_vals)
plt.xlabel('Time')
plt.ylabel('Population Size')
plt.title("Verhulst's Logistic Growth Model (r=0.2, K=1000)" )
plt.show()
```
