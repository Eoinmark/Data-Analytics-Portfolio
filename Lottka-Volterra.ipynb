```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt     #for plotting
from scipy.integrate import odeint  #library for solving ordinary differential equations
```

```python
y0 = [10,10] # [prey, predator] #Initial number of prey and predator in hundreds, 1000 prey and 100 predator

t = np.linspace(0,50,num=1000)


#Initialize the parameters and variables
r = 1.1        #intrinsic rate of prey population increase 
a = 0.4        #predation rate coefficient
b = 0.1        #reproduction rate of predators per 1 prey eaten 
m = 0.4        #predator mortality rate  

# steady state initial conditions
# y0 = [gamma/delta , alpha/beta] # [fish, bears] 


params = [r, a, b, m]

def sim(variables, t, params):    #defining a new variable for the odeint

    # prey population level
    H = variables[0]

    # predator population level
    P = variables[1]


    r = params[0]
    a = params[1]
    b = params[2]
    m = params[3]

#Differential equation for Lotka-Volterra Model
    dHdt = (r * H )- ( a * H * P )      #For the Prey
    dPdt = ( b * H * P ) - ( m * P )    #For the Predator

    return([dHdt, dPdt]) #return the ODEs


y = odeint(sim, y0, t, args=(params,))
```

```python
#plotting our data

f,(ax1,ax2) = plt.subplots(2)

line1, = ax1.plot(t,y[:,0], color="b")



line2, = ax2.plot(t,y[:,1], color="r")

ax1.set_ylabel("Prey (hundreds)")
ax2.set_ylabel("Hunter (hundreds)")
ax2.set_xlabel("Time")

plt.show()
```
