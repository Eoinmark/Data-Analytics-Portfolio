### Introduction

In an attempt to understand as to how predictions of weather pattern at a very far out time is inherently difficult, the American Mathematician Edward Lorenz developed a model, which is a drastic simplification of the weather system. With the help of Margaret Hamilton and Ellen Fetter, they collaboratively developed a simplified mathematical model for atmospheric convection[1].  

To model the atmosphere, which can be considered to exhibit behavior similar to a fluid, they considered a uniformly warmed 2-dimensional fluid from below and cooled above. In this system, convection rolls happens when its temperature is slowly increased (see figure below).

Three quantities characterize this state of the fluid. These mentioned three quantities are: 
<p style="text-align: center;">x: The rate of convective motion <br /> y: The distortion (from linearity) of the vertical temperature profile <br /> z: The distortion (from linearity) of the vertical temperature profile, Z <br /> </p>  

Lorenz then related these three quantities through a system of ordinary differential equations now known as the Lorenz System. This notebook aims to explain the “butterfly effect”, write a python code to solve the Lorenz equation and then analyze the effect of the initial conditions and the system parameters to the behavior of the Lorenz system.



### Theory
The Lorenz equations can be written as follows:


$$\begin{align}
\frac{dx}{dt} = \sigma(y-x) \\
\frac{dy}{dt} = x(\rho - z)-y \\
\frac{dz}{dt} = xy - \beta z \\
\end{align} $$

where x, y, and z are state variables, t is time, σ, ρ, and β are system parameters, and dx/dt, dy/dt, and dz/dt are the rates of change of the state variables over time. <br />

The Lorenz attractor is a strange attractor that emerges when the Lorenz system is plotted in phase space. The attractor has a distinctive butterfly shape that is due to the periodicity of the solutions [1]. The Lorenz attractor is an example of chaos, which is characterized by sensitive dependence on initial conditions. This means that small changes in the initial conditions of the system can lead to vastly different outcomes over time. This is often referred to as the butterfly effect, which was coined by Lorenz himself. The butterfly effect has significant implications for complex systems, as it suggests that small changes can have large effects over time, making accurate long-term predictions difficult or impossible [2]. <br />

Lorenz system, Lorenz attractor, and butterfly effect has been extensively studied and applied in various fields. For example, in atmospheric science, the Lorenz system is used to model atmospheric dynamics and predict weather patterns. In economics, chaos theory has been applied to study financial markets and the behavior of economic systems. In neuroscience, chaos theory has been used to study the dynamics of the brain and its relationship to cognition [1].


### Solving the Lorenz Equations

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt     #for plotting
from scipy.integrate import odeint  #library for solving ordinary differential equations
from mpl_toolkits.mplot3d.axes3d import get_test_data

#Values for different parameters
sig = 10.0
rho = 28.0
beta = 8.0/3

def shm(u,t):
    x = u[0]
    y = u[1]
    z = u[2]
    
    dxdt = sig*(y-x)
    dydt = x*(rho -z)-y
    dzdt = x*y - beta*z
    return [dxdt, dydt, dzdt]
u0 = [0, 1.0, 0]
t =np.linspace(0,100,10000)

sol = odeint (shm,u0,t)

x, y , z = sol[:,0], sol[:,1], sol[:,2]

fig = plt.figure(figsize=plt.figaspect(0.3))
ax = fig.add_subplot(111 , projection='3d')
ax.plot(x, y, z, lw=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Lorenz Attractor')
plt.show()
```

<p>
    The parameter σ controls the rate of convective heat transfer in the Lorenz attractor. From the figure above, increasing σ leads to an increase in the number of folds and twists in the attractor, while decreasing σ reduces the number of folds and twists. A higher value of σ results in a more complex and intricate attractor, while a lower value of σ results in a simpler, less complex attractor [1]. 
The parameter σ represents the Prandtl number, which is the ratio of momentum diffusivity to thermal diffusivity. It is a dimensionless quantity that is used to predict the onset of turbulence in fluid flows. A higher value of σ indicates a more turbulent flow, while a lower value of σ represents a more laminar flow [2]. 
    </p>

<p>The parameter β represents the ratio of the horizontal width of the fluid layer to its height, and it controls the stretching and folding of the Lorenz attractor. Increasing β results in a more stretched and elongated attractor, while decreasing β results in a more compact and compressed attractor [2]. </p>

<p>
In the real world, the parameter β represents the ratio of the Coriolis force to the buoyancy force in a fluid system. A higher value of β indicates a stronger Coriolis force, which leads to more rotation and a more complex flow pattern. A lower value of β represents a weaker Coriolis force, which results in less rotation and a simpler flow pattern [3].</p>


### Sensitive dependence

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt     #for plotting
from scipy.integrate import odeint  #library for solving ordinary differential equations
from mpl_toolkits.mplot3d.axes3d import get_test_data

#Values for different parameters
rho = 28
sig = 10.0
beta = 8.0/3


u0 = [1.0, 1.0, 1.0]

t =np.linspace(0,100,10000)

def shm1(u,t):
    x = u[0]
    y = u[1]
    z = u[2]
    
    dxdt = sig*(y-x)
    dydt = x*(rho -z)-y
    dzdt = x*y - beta*z
    return [dxdt, dydt, dzdt]

sol = odeint (shm1,u0,t)

x1, y1 , z1 = sol[:,0], sol[:,1], sol[:,2]

u0 = [1.00001, 1.00001, 1.00001]
def shm2(u,t):
    x = u[0]
    y = u[1]
    z = u[2]
    
    dxdt = sig*(y-x)
    dydt = x*(rho -z)-y
    dzdt = x*y - beta*z
    return [dxdt, dydt, dzdt]

sol = odeint (shm2,u0,t)

x2, y2 , z2 = sol[:,0], sol[:,1], sol[:,2]


fig = plt.figure(figsize=plt.figaspect(0.3))

ax = fig.add_subplot(1, 3, 1, projection='3d')
ax.plot(x1, y1, z1, lw=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Lorenz Attractor, x=y=z=1')

ax = fig.add_subplot(1, 3, 2, projection='3d')
ax.plot(x2, y2, z2, lw=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Lorenz Attractor, x=y=z=1.00001')

plt.show()
```

<p style="text-align: center;">Figure 7. Solutions to the Lorenz Equation for different initial conditions </p>

<p>
    To understand the sensitive dependence of the Lorenz System to initial conditions, two solutions for the Lorenz system were simulated and generated with different values on the parameters x, y and z. For the plot in the left of figure 7, the initial values for x, y and z are equal to 1, while the right plot has 1.00001 as the initial value for these parameters. As we can observe, the trajectory for both plots in figure 7 is the same initially. However, as the time increases, the trajectory diverges exponentially. From this, the system shows very sensitive dependence on initial conditions since the figure above diverged so much as we increase time even if the difference in the its initial points is only about $10^-5$ in the x, y and z coordinate.  </p>
    
   <p>Such exhibited behavior of the Lorenz attractor is closely related to the chaos theory's <b>butterfly effect</b>, which corresponds to a system's sensitive dependence on initial condition, wherein small differences in one state of a system results to a larger difference in latter states of a nonlinear deterministic system such as the Lorenz System. The term comes from the American Mathematician and Meteorologist, Edward Lorenz' proposition in his presentation before the American Association for the Advancement of Science saying that: </p>
   "If a single flap of a butterfly’s wing can be instrumental in generating a tornado, so also can all he previous and subsequent flaps of its wings, as can the flaps of the wings of millions of other butterflies, not to mention the activities of innumerable more powerful creatures, including our own species." [4]

The implication of this observation about the sensitive dependence of such systems on initial conditions is how it is impossible to predict future states of chaotic systems, as the prediction breaks down as we go far out in time due to how these system reacts on the slightest of change in its earlier states [1] .
