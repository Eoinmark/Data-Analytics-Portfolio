# Logistic Map

### About this project
A logistic map is a mathematical function used in population growth modeling. It is a difference equation that depicts the evolution of a population over time in a closed system [1]. It is defined by the recurrence relation given by,

$$
\begin{align}
x_{n+1} = r \bullet x_{n} \bullet (1-x_{n})
\end{align}
$$

Where $x_{n}$ represents the population at some time, $n$ and $r$ represents the growth rate of the population. The aim of this project is to explore the dependence of the behavior of the logistic map on different values of these parameters using python.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Initialize seed and growth parameter, r
N = 50 #number of iterations
x = 0.2+np.zeros(N) #for {xn}
x_prime = 0.200001+np.zeros(N)  #seed 1
x_diff = np.zeros(N)            #seed 2
r = 2


#Calculate for terms using logistic map equation
for n in range(N-1):
    x[n+1] = r*x[n]*(1-x[n])
    x_prime[n+1] = r*x_prime[n]*(1-x_prime[n])
    x_diff = abs(x - x_prime)
    
plt.figure()
plt.plot(x, 'r', label= '$x_0 = 0.2 $')
plt.plot(x_prime, '--b', label='$x_0 = 0.200001$')
plt.title('Logistic Map, r=2')
plt.xlabel('N')
plt.ylabel('x')
plt.legend()
plt.show
```
